Matplotlib is a Python 2D plotting library which produces quality figures in a variety of hardcopy formats and interactive environments. Matplotlib can be used in Python scripts, the Python and IPython shell, the jupyter notebook, web application servers, and four graphical user interface toolkits. 

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

%matplotlib is a magic function in IPython. With this backend, the output of plotting commands is displayed inline within frontends like the Jupyter notebook, directly below the code cell that produced it. The resulting plots will then also be stored in the notebook document. 

```python
plt.plot(np.array([1,2,3]),np.array([1,2,3]))
```

```python
x = np.array([1,2,3])
plt.plot(x,x**3,label = 'cubed')
plt.plot(x,x**5, label = 'raised to 5')
plt.legend()
```

## Subplots 

```python
plt.subplot(1,3,1)
plt.plot(x,x**2,label = 'squared')
plt.legend()
plt.subplot(1,3,2)
plt.plot(x,x**3, label = 'cubed',color = 'red')
plt.legend()
plt.subplot(1,3,3)
plt.plot(x,x**4, label = 'raised to 4',color = 'green')
plt.legend()
```

# Seaborn-Statistical plot 

Seaborn is a Python visualization library based on matplotlib. It provides a high-level interface for drawing attractive statistical graphics.

```python
import seaborn as sns
```

```python
sns.distplot(df['SepalWidthCm'])
```

The most convenient way to take a quick look at a univariate distribution( distribution of single variable) in seaborn is the distplot() function. By default, this will draw a histogram. 

```python
sns.distplot(df['PetalLengthCm'])
```

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data = df)
```

Join plot returns the scatter plot of 2 variable as well as distribution of each variable

```python
sns.pairplot(Iris.drop(['Id'],axis=1),hue = 'Species')
```

Pairplot draw scatterplots for joint relationships and histograms for univariate distributions in diagonal.

## Categorical plots 

```python
sns.barplot(x='Species',y='SepalLengthCm',data=df)
```

A bar plot represents the average for a numeric variable with the height of each rectangle.

```python
sns.countplot(x='Species',data=Iris)
```

Countplot is a barplot where the dependent variable is the number of instances of each instance of the independent variable.
