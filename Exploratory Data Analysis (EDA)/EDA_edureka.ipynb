## Cleaning the data 

```python
students.isnull().sum()
```

### Dropping unwanted columns 

```python
students = students.drop(["race/ethnicity", "parental level of education"], axis=1)
```

```python
students.head()
```

## Relationship analysis 

```python
corelation = students.corr()
```

```python
sns.heatmap(
    corelation, xticklabels=True, yticklabels=True, annot=True,
)
```

```python
sns.pairplot(students)
```

```python
cols = ["gender", "lunch", "test preparation course"]

for col in cols:
    sns.relplot(x="math score", y="reading score", data=students, hue=students[col])
```

```python
sns.distplot(students["math score"], bins=5)
```

```python
cols = ["math score", "reading score", "writing score"]

for col in cols:
    sns.catplot(x=col, kind="box", data=students)
```

This helps in visualizing the outliers.
