```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
climate = pd.read_csv('climate_change.csv', parse_dates = ['Date'], index_col = 'Date')
climate.head()
```

### Plotting time series 

```python
fig,ax = plt.subplots()
ax.plot(climate.index,climate['Interpolated'])
ax.set_xlabel('Year')
ax.set_ylabel('CO2 concentration in ppm')
plt.show()
```

### Zooming in 

```python
nineties_onwards = climate['1990-01-01':'2020-12-31']
fig,ax = plt.subplots()
ax.plot(nineties_onwards.index, nineties_onwards['Interpolated'])
ax.set_xlabel('Year')
ax.set_ylabel('CO2 concentration in ppm')
plt.show()
```

### Plotting 2 variables 

```python
fig,ax = plt.subplots()

ax.plot(climate.index,climate['Interpolated'],color = 'r')
ax.set_xlabel('Year')
ax.set_ylabel('Interpolated CO2 concentration (ppm)',color = 'r')
ax.tick_params('y',colors = 'r')

ax2 = ax.twinx()

ax2.plot(climate.index,climate['Average'],color = 'b')
ax2.set_ylabel('Average CO2 concentration (ppm)',color = 'b')
ax2.tick_params('y',colors = 'b')

plt.show()
```

### Function creation 

```python
def timeplot(axes, x, y, color, xlabel, ylabel):
    axes.plot(x,y,color = color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color = color)
    axes.tick_params('y', colors = color)
```

```python
fig,ax = plt.subplots()
timeplot(ax,climate.index,climate['Interpolated'],'r','Year','Interpolated CO2 concentration (ppm)')

ax2 = ax.twinx()

timeplot(ax2,climate.index,climate['Average'],'b','Year','Average CO2 concentration (ppm)')

plt.show()
```
