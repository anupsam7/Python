```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gold = pd.read_csv("gold_top5.csv", index_col=0)
silver = pd.read_csv("silver_top5.csv", index_col=0)
bronze = pd.read_csv("bronze_top5.csv", index_col=0)
```

## Bar charts 

```python
fig, ax = plt.subplots()
ax.bar(gold.index, gold["Total"])
ax.set_xticklabels(gold.index, rotation=90)
ax.set_ylabel("Number of medals")
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(gold.index, gold["Total"], label="Gold")
ax.bar(gold.index, silver["Total"], bottom=gold["Total"], label="Silver")
# ax.bar(gold.index,bronze['Total'],bottom = gold['Total'] + silver['Total'], label = 'Bronze')

ax.set_xticklabels(gold.index, rotation=90)
ax.set_ylabel("Number of medals")

ax.legend()
plt.show()
```

## Histogram 

```python
fig, ax = plt.subplots()

ax.hist(gold["Total"], bins=10, label="Gold")
ax.hist(silver["Total"], bins=10, label="Silver")
ax.hist(bronze["Total"], bins=10, label="Bronze")

ax.set_xlabel("# of medals")
ax.set_ylabel("# of observations")

ax.legend()
plt.show()
```

### Transparent Histograms 

```python
fig, ax = plt.subplots()

ax.hist(gold["Total"], bins=10, label="Gold", histtype="step")
ax.hist(silver["Total"], bins=10, label="Silver", histtype="step")
ax.hist(bronze["Total"], bins=10, label="Bronze", histtype="step")

ax.set_xlabel("# of medals")
ax.set_ylabel("# of observations")

ax.legend()
plt.show()
```

## Error Bars 

```python
fig, ax = plt.subplots()

ax.bar("Gold", gold["Total"].mean(), yerr=gold["Total"].std())
ax.bar("Silver", silver["Total"].mean(), yerr=silver["Total"].std())
ax.bar("Bronze", bronze["Total"].mean(), yerr=bronze["Total"].std())

ax.set_xlabel("Medal Type")
ax.set_ylabel("Mean # of Medals")

plt.show()
```

The error bars show the variation of values within each medal type.

### Adding error bars to line plots 

```python
pitts = pd.read_csv("pittsburgh2013.csv")
pitts1 = pitts[:20]
pitts.head()
```

```python
fig, ax = plt.subplots()

ax.errorbar(
    pitts1["Date"],
    pitts1["Max TemperatureF"],
    yerr=pitts1["Max TemperatureF"].std(),
    label="MaxTemp(F)",
)
ax.errorbar(
    pitts1["Date"],
    pitts1["Min TemperatureF"],
    yerr=pitts1["Min TemperatureF"].std(),
    label="MinTemp(F)",
)

ax.set_xticklabels(pitts1["Date"], rotation=90)
ax.set_xlabel("Date")
ax.set_ylabel("Temperature in F")
ax.set_title("Max and Min Temperatures for Pittsburgh")

ax.legend()
plt.show()
```

## Boxplots 

```python
fig, ax = plt.subplots()

ax.boxplot([gold["Total"], silver["Total"]])

ax.set_xticklabels(["Gold", "Silver"])
ax.set_ylabel("Number of medals")

plt.show()
```

## Scatter plots 

```python
fig,ax = plt.subplots()

ax.scatter(pitts1['Max TemperatureF'],pitts1['Max Humidity'])

ax.set_xlabel('Max Temperature (F)')
ax.set_ylabel('Max Humidity (%)')

plt.show()
```
