```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
seattle = pd.read_csv('seattle_weather.csv')
seattle.head()
```

```python
austin = pd.read_csv('austin_weather.csv')
austin.head()
```

```python
seattle1 = seattle[:20]
austin1 = austin[:20]
```

```python
fig,ax = plt.subplots()
ax.plot(seattle1['date'],seattle1['temp_max'])
ax.plot(austin1['Date'],austin1['TempHighF'])
plt.show()
```

Note: The temperature for Austin is in Fahrenheit while that for Seattle is in Celsius.

### Markers, Linestyle & Colour 

```python
fig,ax = plt.subplots()
ax.plot(seattle1['date'],seattle1['temp_max'],marker = 'v', linestyle = '--', color = 'b')
ax.plot(austin1['Date'],austin1['TempHighF'], marker = 's', linestyle = '--', color = 'r')
plt.show()
```

```python
seattle1.plot('date','temp_max',marker = 's',linestyle = 'None', color = 'r')
plt.show()
```

### Axis Labels & Title 

```python
fig,ax = plt.subplots()
ax.plot(seattle1['date'],seattle1['temp_max'],marker = 'v', linestyle = '--', color = 'b')
ax.plot(austin1['Date'],austin1['TempHighF'], marker = 's', linestyle = '--', color = 'r')
ax.set_xlabel('Date')
ax.set_ylabel('Max Temp')
ax.set_title('Max temp comparisons for Seattle & Austin')
plt.show()
```

### Sub-plots 

```python
fig,ax = plt.subplots(2,1)
plt.show()
ax.shape
```

```python
fig,ax = plt.subplots(2,1)
ax[0].plot(seattle1['date'],seattle1['temp_max'],marker = 'v', linestyle = '--', color = 'b')
ax[1].plot(austin1['Date'],austin1['TempHighF'], marker = 's', linestyle = '--', color = 'r')
ax[0].set_ylabel('Max Temp in C')
ax[1].set_ylabel('Max Temp in F')
ax[0].set_xlabel('Date')
plt.show()
```

### Sharing the same y-axis range 

```python
fig,ax = plt.subplots(2,1,sharey = True)
ax[0].plot(seattle1['date'],seattle1['temp_max'],marker = 'v', linestyle = '--', color = 'b')
ax[1].plot(austin1['Date'],austin1['TempHighF'], marker = 's', linestyle = '--', color = 'r')
ax[0].set_ylabel('Max Temp in C')
ax[1].set_ylabel('Max Temp in F')
ax[0].set_xlabel('Date')
plt.show()
```
