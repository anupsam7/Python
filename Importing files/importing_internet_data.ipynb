## HTTP requests to import files 

```python
import requests
```

```python
# specifying the url
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"

# Packaging the request, sending the request and catching the response
r = requests.get(url)

# Extracting the response
text = r.text

print(text)
```

## Scraping the web using BeautifulSoup

```python
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"
r = requests.get(url)
html_doc = r.text

# Creating the BeautifulSoup object
soup = BeautifulSoup(html_doc)
```

```python
print(soup.prettify())
```

```python
# printing the title of the webpage
soup.title
```

```python
# printing the text
soup.get_text()
```

```python
# finding and printing all the hyperlinks in the article
for link in soup.find_all("a"):
    print(link.get("href"))
```
