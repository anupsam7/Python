```python
import numpy as np
import pandas as pd
```

```python
gold = pd.read_csv('gold_top5.csv',index_col=0)
bronze = pd.read_csv('bronze_top5.csv',index_col=0)
```

```python
pd.concat([gold,bronze],axis=1,join='inner')
```

### Merging 

```python
pd.merge(gold,bronze,on = 'Country')
```

```python
pd.merge(gold,bronze,on = 'Country',suffixes = ['_gold','_bronze'])
```

### Left & Right Merges 

```python
pd.merge(gold,bronze,on = 'Country',suffixes = ['_gold','_bronze'],how = 'left')
```

```python
pd.merge(gold,bronze,on = 'Country',suffixes = ['_gold','_bronze'],how = 'right')
```

### Ordered merges 

```python
software = pd.read_csv('feb-sales-Software.csv',parse_dates = ['Date']).sort_values('Date')
hardware = pd.read_csv('feb-sales-Hardware.csv',parse_dates = ['Date']).sort_values('Date')
```

```python
pd.merge(hardware,software,how = 'outer').head()
```
