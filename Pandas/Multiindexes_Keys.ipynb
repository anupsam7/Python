## Multi-indexing 

```python
import pandas as pd
gold = pd.read_csv('Gold.csv', index_col = 0)
silver = pd.read_csv('Silver.csv', index_col = 0)
```

```python
pd.concat([gold,silver],keys = ['gold','silver'],axis=1)
```

```python
pd.concat([gold,silver],keys = ['gold','silver'],axis=0)
```

### pd.concat() with dict 

```python
medal_dict = {'gold':gold, 'silver':silver}
```

```python
pd.concat(medal_dict, axis = 1)
```

```python
pd.concat(medal_dict, axis = 0)
```

### Concatenating all medals 

```python
medals = []
types = ['gold','silver','bronze']
medal_type = ['gold','silver','bronze']
```

```python
for medal in medal_type:
    
    file_name = '%s_top5.csv' % medal
    
    medal_df = pd.read_csv(file_name, index_col = 'Country')
    
    medals.append(medal_df)
    
medals = pd.concat(medals, keys = types)

print(medals)
```
