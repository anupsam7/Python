### Un-pickle the data

```python
import pickle

with open('homeless_data.pkl', 'rb') as f:
    data = pickle.load(f)
```

### Data exploration 

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

### Histograms 

```python
data['region'].hist()
plt.show()
```

```python
bins=[10,20]
for b in bins:
    data['state_pop'].hist(bins=b)
    plt.show()
```

### Layering plots 

```python
data[data['region']=='Pacific']['state_pop'].hist(bins=10,alpha=0.7)
data[data['region']=='Mountain']['state_pop'].hist(bins=10, alpha = 0.7)
plt.legend(['Pacific','Mountain'])
plt.show()
```

The alpha attribute helps in visualizing layered plots. A value of 0 means completely transparent and 1 means compeletely opaque.

### Bar-plots 

Shows the relation between a categorical variable and a numerical variable.

```python
avg_pop_region = data.groupby('region')['state_pop'].mean()
avg_pop_region.plot(kind = 'bar', title = 'Average population by region')
plt.show()
```

```python
with open('walmart_sales.pkl', 'rb') as f:
    data1 = pickle.load(f)
```

```python
data1.head()
```

### Line plots 

To visualize the variation of a continuous variable over time.

```python
data1.plot(x='date',y='fuel_price_usd_per_l',kind='line',rot=45)
plt.show()
```

The 'rot' argument of the plot method rotates the x-axis line markings by the given angle.

### Scatter plots 

Great for visualizing relationship between 2 numerical variables.

```python
data1.plot(x='temperature_c',y = 'weekly_sales',kind='scatter')
plt.show()
```
