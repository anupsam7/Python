### Un-pickle the data 

```python
import pickle

with open('homeless_data.pkl', 'rb') as f:
    data = pickle.load(f)
```

### Data exploration 

```python
data.head()
```

### .pivot_table() method 

```python
data.pivot_table(values = 'state_pop',index = 'region')
```

By default, pivot table calculates the mean.

```python
import numpy as np
data.pivot_table(values='state_pop',index = 'region', aggfunc = [max, min, np.median, np.mean] )
```

### Pivoting on 2 variables 

```python
data.pivot_table(values = 'state_pop', index = 'region', columns = 'state')
```

```python
data.pivot_table(values = 'state_pop', index = 'region', columns = 'state', fill_value = 0) # fills missing values with 0
```

### Summing pivot tables 

```python
data.pivot_table(values = 'state_pop', index = 'region', columns = 'state', fill_value = 0, margins = True)
# gives the mean value of all the rows and columns
```
