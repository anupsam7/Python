```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
mpg = sns.load_dataset("mpg")
mpg.head()
```

## Plot of each observation 

### Strip plot 

```python
sns.stripplot(data=mpg, x="origin", y="horsepower", jitter=True)
plt.show()
```

### Swarm plot 

```python
sns.swarmplot(data=mpg, x="origin", y="horsepower")
plt.show()
```

## Abstract Representation 

### Box plot 

```python
sns.boxplot(data=mpg, x="origin", y="displacement")
plt.show()
```

### Violin plot 

```python
sns.violinplot(data=mpg, x="origin", y="displacement", palette="husl")
plt.show()
```

### boxen plot (previously called lv plot)

```python
sns.boxenplot(data=mpg, x="origin", y="displacement", palette="Paired")
plt.show()
```

## Statistical estimates 

### Bar plot 

```python
sns.barplot(data=mpg, x="origin", y="mpg")
plt.show()
```

```python
sns.barplot(data=mpg, x="origin", y="mpg", hue="cylinders")
plt.show()
```

### Point plot 

```python
sns.pointplot(data=mpg, x="origin", y="mpg", capsize=0.1)
plt.show()
```

### Countplot 

```python
sns.countplot(data=mpg, x="origin")
plt.show()
```

## Regression plots 

```python
diamond = sns.load_dataset("diamonds")
diamond.head()
```

```python
sns.regplot(
    x="carat", y="price", data=diamond, marker="+"
)  # fit_reg = False as an argument to regplot will disable the regression line
plt.show()
```

```python
sns.residplot(x="carat", y="price", data=diamond)
plt.show()
```

A residual plot is useful for evaluating the fit of the model. This is used to determine if there is a linear relation between the 2 variables.

### Polynomial Regression 

```python
sns.regplot(x="carat", y="price", data=diamond, order=2, color="g")
plt.show()
```

### Regression with categorical values 

```python
sns.regplot(x="cylinders", y="horsepower", data=mpg, order=2, x_jitter=0.1)
plt.show()
```

'cylinders' is a categorical variable having numerical variable and hence linear regression can be performed. 

### Estimators 

```python
sns.regplot(
    x="cylinders", y="horsepower", data=mpg, order=2, x_estimator=np.mean, color="r"
)
plt.show()
```

### Binning the data 

```python
sns.regplot(x="horsepower", y="mpg", data=mpg, x_bins=4, color="r")
plt.show()
```

x_bins can be used to divide the continuous data into discrete bins. 

## Matrix plots 

```python
daily_show = pd.read_csv("daily-show-guests.csv")
daily_show.head()
```

```python
pd_tab = pd.crosstab(daily_show["Group"], daily_show["YEAR"])
print(pd_tab)
```

```python
sns.heatmap(pd_tab)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.heatmap(pd_tab, cbar=False, cmap="YlGnBu", linewidths=0.5) # cbar = True shows the color index
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```

### Correlation heatmap 

```python
sns.heatmap(mpg.corr())
plt.show()
```
