```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
student = pd.read_csv("student-alcohol-consumption.csv", index_col=0)
student.head()
```

## Relative plots 

```python
sns.relplot(x="absences", y="G1", data=student, kind="scatter")
plt.show()
```

### Subplots by col 

```python
sns.relplot(x="absences", y="G1", data=student, kind="scatter", col="location")
plt.show()
```

There are more urban students and therefore greater number of absentee students.

### Subplots by row 

```python
sns.relplot(x="absences", y="G1", data=student, kind="scatter", row="sex")
plt.show()
```

### Wrapping Columns 

```python
sns.relplot(
    x="absences", y="G1", data=student, kind="scatter", col="study_time", col_wrap=2
)
plt.show()
```

The number of absences is the least among students wo study for more than 10 hours.

## Customizing scatter plots 

```python
mpg = pd.read_csv('mpg.csv',index_col=0)
mpg.head()
```

```python
sns.relplot(x='weight',y='horsepower',data=mpg,kind='scatter')
plt.show()
```

### Calculation of correlation 

```python
mpg.isnull().sum()
```

```python
mpg = mpg.dropna(axis=0)
```

```python
np.corrcoef(mpg['weight'],mpg['horsepower'])
```

The weight of the vehicle is strongly correlated to its mpg with a R value of 0.86.

### Point size, hue and style 

```python
sns.relplot(x = 'weight', y = 'acceleration', data = mpg, hue = 'origin', style = 'origin')
plt.show()
```

American cars tend to be heavier while European cars tend to have higher acceleration.

```python
sns.relplot(x = 'weight', y = 'acceleration', data = mpg, hue = 'cylinders', size = 'cylinders')
plt.show()
```

The number of cylinders is directly proportional to weight and indirectly proportional to acceleration.

## Line plots 

```python
sns.relplot(x = 'model_year', y = 'horsepower', data = mpg, kind = 'line')
plt.show()
```

The overall horsepower of vehicles have been descreasing over the years

### Replacing confidence interval with standard deviation 

```python
sns.relplot(x = 'model_year', y = 'horsepower', data = mpg, kind = 'line', ci = 'sd')
plt.show()
```

### Country-wise comparison 

```python
sns.relplot(x = 'model_year', y = 'horsepower', data = mpg, kind = 'line', style = 'origin', hue = 'origin', ci = None)
plt.show()
```

```python
sns.relplot(x = 'model_year', y = 'horsepower', data = mpg, kind = 'line', style = 'origin', hue = 'origin', ci = None, markers = True, dashes = False)
plt.show()
```

There has been a more distinct decrease in the horsepower of American cars as compared to Japanese and European.
