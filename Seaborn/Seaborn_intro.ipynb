```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
countries = pd.read_csv("countries-of-the-world.csv", index_col=0)
countries.head()
```

## Scatter plots 

```python
sns.scatterplot(countries["Population"], countries["GDP ($ per capita)"])
plt.show()
```

```python
sns.scatterplot(countries["GDP ($ per capita)"], countries["Literacy (%)"])
plt.yticks(rotation=90)
plt.show()
```

## Count plots 

```python
sns.countplot(countries["Region"])
plt.xticks(rotation=90)
plt.show()
```

## Adding a third variable with hue 

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(tips["total_bill"], tips["tip"], hue=tips["sex"])
plt.show()
```

```python
sns.countplot(tips["day"], hue=tips["smoker"])
plt.show()
```

### Defining color palette 

```python
color_palette = {"Yes": "Red", "No": "Green"}
sns.countplot(tips["day"], hue=tips["smoker"], palette=color_palette)
plt.show()
```

### Setting hue order 

```python
color_palette = {"Yes": "Red", "No": "Green"}
sns.countplot(
    tips["day"], hue=tips["smoker"], palette=color_palette, hue_order=["No", "Yes"]
)
plt.show()
```
