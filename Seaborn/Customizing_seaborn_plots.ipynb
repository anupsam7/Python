```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
young = pd.read_csv("young-people-survey-responses.csv", index_col=0)
young.head()
```

```python
sns.countplot(young["Internet usage"])
plt.show()
```

## Changing Figure Style 

```python
sns.set_style("darkgrid")
sns.countplot(young["Internet usage"])
plt.show()
```

```python
sns.set_style("whitegrid")
sns.countplot(young["Internet usage"])
plt.show()
```

## Changing the Palette 

```python
sns.set_style("whitegrid")
sns.set_palette("RdBu")
sns.countplot(young["Internet usage"])
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_palette("PRGn")
sns.countplot(young["Internet usage"])
plt.show()
```

## Changing the scale 

```python
sns.set_style("whitegrid")
sns.set_palette("deep")
sns.countplot(young["Internet usage"])
plt.show()
```

This is the default context of paper. The graph below switches it to the larger context of talk.

```python
sns.set_style("whitegrid")
sns.set_palette("deep")
sns.set_context("talk")
sns.countplot(young["Internet usage"])
plt.xticks(rotation=45)
plt.show()
```

## Titles & Labels 

### FacetGrid vs AxesSubplots  

Both are objects created by Seaborn plots. FacetGrid includes relplot and catplot, which can create subplots. AxesSubplot includes scatterplot and countplot that create only a single plot 

### FacetGrid 

```python
countries = pd.read_csv('countries-of-the-world.csv', index_col=0)
countries.head()
```

```python
g = sns.catplot(x = 'Region', y = 'GDP ($ per capita)', data=countries, kind='box')
sns.set_context('paper')
plt.show()
```

```python
g = sns.catplot(x = 'Region', y = 'GDP ($ per capita)', data=countries, kind='box')
g.fig.suptitle('Analysis of GDP per capita', y = 1.03)
plt.xticks(rotation = 90)
plt.show()
```

For AxesSubplots, use g.set_title instead of g.fig.suptitle ;
Also g.set(xlabel = "", ylabel = "")
