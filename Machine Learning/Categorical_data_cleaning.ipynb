```python
import numpy as np
import pandas as pd
```

## Read in data 

```python
titanic = pd.read_csv('Titanic.csv')
```

```python
titanic.drop(['Name','Ticket'],axis = 1, inplace = True)
```

```python
titanic.head()
```

## Creating cabin indicator 

```python
titanic['Cabin_int'] = np.where(titanic['Cabin'].isnull(),0,1)
```

```python
titanic.head()
```

## Converting Sex column to numerical values 

```python
Sex_num = {'male':0,'female':1}
titanic['Sex'] = titanic['Sex'].map(Sex_num)
```

```python
titanic.head()
```

## Dropping Embarked and Cabin columns 

```python
titanic.drop(['Embarked','Cabin'],axis = 1, inplace = True)
```
